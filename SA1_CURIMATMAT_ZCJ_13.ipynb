**13.** Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average.

```python
import time
import random
import matplotlib.pyplot as plt
import math

def experimental():
    sizes = [1000, 2000, 5000, 10000, 20000, 50000, 100000]
    times = []

    for size in sizes:
        data = random.sample(range(size * 10), size)
        start = time.time()
        sorted(data)
        times.append(time.time() - start)

    plt.plot(sizes, times, label='Actual Time', marker='o')

    n_log_n = [size * math.log(size) for size in sizes]
    normalized = [x * (times[-1] / n_log_n[-1]) for x in n_log_n]

    plt.plot(sizes, normalized, label='O(n log n)', linestyle='--')

    plt.xlabel('List Size (n)')
    plt.ylabel('Average Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

experimental()
```

**P- 4.23** Implement a recursive function with signature find (path, filename) that reports all entries of the file system rooted at the given path having the given filename.

```python
import os
from google.colab import files

def find(path, filename):
    if not os.path.exists(path):
        print(f"The path '{path}' does not exist.")
        return

    for root, dirs, files in os.walk(path):
        if filename in files:
            print(f"Found: {os.path.join(root, filename)}")
            return
    print(f"'{filename}' not found in '{path}'.")

print("Please upload your files:")
uploaded = files.upload()
search_filename = input("Enter the filename to search for: ")
find('/content', search_filename)
```

**P-4.25** Provide a nonrecursive implementation of the draw_interval function for
the English ruler project of Section 4.1.2. There should be precisely 2c−1
lines of output if c represents the length of the center tick. If incrementing
a counter from 0 to 2c −2, the number of dashes for each tick line should
be exactly one more than the number of consecutive 1’s at the end of the
binary representation of the counter.

```python
def draw_line(tick_length, tick_label=''):
    """Draw one line with given tick length (followed by optional label)."""
    line = '-' * tick_length
    if tick_label:
        line += ' ' + tick_label
    print(line)

def count_trailing_ones(n):
    """Count the number of trailing 1's in the binary representation of n."""
    count = 0
    while (n & 1) == 1:
        count += 1
        n >>= 1
    return count

def draw_interval(center_length):
    """Draw tick interval based upon a central tick length."""
    total_lines = (1 << center_length) - 1

    for i in range(total_lines):
        trailing_ones = count_trailing_ones(i)
        tick_length = trailing_ones + 1
        draw_line(tick_length)

def draw_ruler(num_inches, major_length):
    draw_line(major_length, '0')
    for j in range(1, 1 + num_inches):
        draw_interval(major_length - 1)
        draw_line(major_length, str(j))


draw_ruler(6, 4)

```

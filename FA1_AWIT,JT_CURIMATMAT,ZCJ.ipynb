# **P-1.29 Write a Python program that outputs all possible strings formed by using the characters c , a , t , d , o , and g exactly once**

```python
def permute(str, perm = ""):
    if len(str) == 0:
        print(perm)
    else:
        for index in range(len(str)):
            current_char = str[index]
            remaining = str[:index] + str[index+1:]
            permute(remaining, perm + current_char)

chars = ['c', 'a', 't', 'd', 'o', 'g']

permute(chars)
```
